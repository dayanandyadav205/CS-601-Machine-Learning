```python
#Distance from the origin (Euclidean Formula)
import numpy as np

# define a n-dimensional vector A
A = np.array([1, 2])

# Calculate the Euclidean distance from th origin (L2 Norm)
distance = np.linalg.norm(A) #linalg = Linear Algebra

print("Euclidean distance from origin:", distance)
```

```python
#Distance from the origin (Euclidean Formula) n-D vector
import numpy as np

# define a n-dimensional vector A
A = np.array([1, 2, 3, 4, 5])
B = np.array([6, 7, 8, 9, 10])

#Calculate the Difference Vector
difference = A - B

difference
```

```python
distance = np.linalg.norm(difference)

print("Euclidean distance between A and B:", distance)
```

```python
#Mean Centering
import numpy as np
import matplotlib.pyplot as plt

#Generate random 2D data with 100 points
data = np.random.rand(100, 2)

data
```

```python
#Plot the original data
plt.scatter(data[:, 0], data[:, 1])
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Before Mean Centering")
```

```python
#Calculate the mean of data
data_mean = np.mean(data, axis=0)

print(data_mean)

#perform mean centering
centered_data = data - data_mean
```

```python
#Plot the mean centered data
plt.scatter(centered_data[:, 0], centered_data[:, 1], color="orange")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("After Mean Centering")
```

```python
#Mean Centering
import numpy as np
import matplotlib.pyplot as plt

#Generate random 2D data with 100 points
data = np.random.rand(100, 2)

data

#Plot the original data
plt.scatter(data[:, 0], data[:, 1])
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Before Mean Centering")

#Calculate the mean of data
data_mean = np.mean(data, axis=0)

print(data_mean)

#perform mean centering
centered_data = data - data_mean

#Plot the mean centered data
plt.scatter(centered_data[:, 0], centered_data[:, 1], color="orange")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("After Mean Centering")
```
