```python
import numpy as np

# function to print Checkerboard pattern
def printcheckerboard(n):
    print("Checkerboard pattern:")

    # create a n * n matrix
    x = np.zeros((n, n), dtype=int)

    # fill with 1 the alternate cells in rows and columns
    x[1::2, ::2] = 1 # Odd rows indexes
    x[::2, 1::2] = 1 # odd columns indexes

    # print the pattern
    for i in range(n):
        for j in range(n):
            print(x[i][j], end=" ")
        print()

# driver code
n = 2
printcheckerboard(n)
```
